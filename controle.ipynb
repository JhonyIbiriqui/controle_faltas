```python
import pandas as pd

tb_base = pd.read_csv(r'datasets\controle de faltas.csv')


display(tb_base)
```

```python
Faltas_colab = tb_base.groupby("NOME COLABORADOR") ["TEMPO(DIAS)"].sum().reset_index()

#Faltas_colab = Faltas_colab.sort_values(by="NOME COLABORADOR" ,ascending=False)

substituicoes = {
    "JHONY": "JHONY ALLISON",
    "HUGO": "HUGO MANICARDI",
    "RAFAELA": "RAFAELA LUPPI"
}

Faltas_colab["NOME COLABORADOR"] = Faltas_colab["NOME COLABORADOR"].replace(substituicoes)

Faltas_colab = Faltas_colab.groupby("NOME COLABORADOR") ["TEMPO(DIAS)"].sum().reset_index()

display(Faltas_colab)

```
